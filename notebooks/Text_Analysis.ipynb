```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the Excel file
file_path = '/root/shared/Text_Analysis.xlsx'
```

```python
sheet_name = 'Sheet1'
data = pd.read_excel(file_path, sheet_name=sheet_name, header = 1)

# Extract the years and data for plotting
years = data.columns[1:]  # years are in the columns from B to T

# Plot for "World" only
world_data = data[data['Region, country, or economy'] == 'World'].iloc[0, 1:].astype(int)
plt.figure(figsize=(10, 6))
plt.plot(years, world_data, marker='o')
plt.title('S&E Articles for World (2000-2018)')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.yscale('log')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot for all regions except "World"
regions_data = data[data['Region, country, or economy'] != 'World']
plt.figure(figsize=(12, 8))
for index, row in regions_data.iterrows():
    region_name = row['Region, country, or economy']
    region_data = row[1:].astype(int)
    plt.plot(years, region_data, marker='o', label=region_name)

plt.title('S&E Articles by Region (2000-2018)')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sheet_name = 'Sheet2'

# Read the data, using the first row as header
data = pd.read_excel(file_path, sheet_name=sheet_name, header=0)

# Display the column names to confirm
print(data.columns)

# Extract the years and data for plotting
years = data.columns[1:]  # years are in the columns from B to T

# Plot for each region
plt.figure(figsize=(12, 8))
for index, row in data.iterrows():
    region_name = row['Region, country, or economy']
    region_data = row[1:].astype(int)
    plt.plot(years, region_data, marker='o', label=region_name)

plt.title('S&E Articles by Region, Country, or Economy (2000-2018)')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
